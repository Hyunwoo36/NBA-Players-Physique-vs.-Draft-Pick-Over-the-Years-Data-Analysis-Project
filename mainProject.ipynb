# Abstract
In this study, we explored the evolving trends in NBA player attributes over the past two decades, transitioning from dominant big men to agile modern players. Using datasets spanning the 2000-2001 to 2023-2024 drafts, we analyzed height, weight, and BMI, considering variations across player positions.

Our findings revealed a consistent height over 20 years but a statistically significant decrease in average weight and BMI. Comparing 2019-2023 with 2000-2018, weight and BMI statistically significantly decreased, aligning with expectations. ANOVA tests highlighted variations across player positions, supporting initial hypotheses. Surprisingly, there was no correlation between player positions and draft picks over the last 20 years, remaining independent in the last 5 years as well.

In conclusion, our study unveils nuanced insights into NBA player attributes, confirming some expectations and challenging assumptions. Future research should delve into these intricacies for a comprehensive understanding, acknowledging limitations in the observed variability.


# Research Question

Are there noticeable trends in the NBA over the last 20 years regarding players' height and weight? More specifically, our goal is to answer the following questions:
- Has there been a statistically significant shift in the average height and weight of NBA players over the last five years?
- Is there any correlation between players’ height/weight ratio and their draft pick?
- Is there any correlation between players’ height/weight ratio and their position?

## Background and Prior Work


&emsp; Since the 1990s, the National Basketball Association (NBA) has seen significant changes in player physicality and playing styles, largely due to evolving rules <a name="cite_ref-4"></a>[<sup>4</sup>](https://www.deseret.com/2019/2/21/20666425/nba-rules-have-adapted-over-the-years-to-make-the-game-more-fun-for-players-fans). The earlier era favored robust players like Shaquille O’Neal, emphasizing defense and physical strength. However, recent rule changes have facilitated a faster game pace, resulting in a shift towards leaner player builds <a name="cite_ref-3"></a>[<sup>3</sup>](https://www.topendsports.com/sport/basketball/anthropometry-nba.htm). This evolution is exemplified by players such as Kristaps Porziņģis and Victor Wembanyama, who combine height with a leaner physique​. 

&emsp; The NBA categorizes players into five positions: point guards (PG), shooting guards (SG), small forwards (SF), power forwards (PF), and centers (C). Each role has distinct responsibilities on the court, ranging from offensive playmaking and scoring to rebounding and defense. This classification system has been instrumental in shaping team strategies and player development​ <a name="cite_ref-5"></a>[<sup>5</sup>](https://jr.nba.com/basketball-positions/).

&emsp; The NBA Draft is a critical process for team roster formation and league competitiveness. Teams select eligible players over two rounds, with the first round order partly determined by a lottery system favoring teams with poorer records. This method aims to maintain competitive balance by giving lower-performing teams access to higher-talented prospects​ <a name="cite_ref-6"></a>[<sup>6</sup>](https://www.nba.com/news/nba-draft-lottery-explainer).

&emsp; The NBA's recent emphasis on speed and shooting accuracy has led to players focusing more on developing leaner muscle builds<a name="cite_ref-1"></a>[<sup>1</sup>](https://harvardsportsanalysis.org/2017/11/maybe-we-should-call-it-skinnyball-weight-vs-height-in-the-nba/#:~:text=The%20weight%20of%20the%20league,entirety%20of%20the%2021%20century). This trend is evident in teams like the Golden State Warriors, <a name="cite_ref-3"></a>[<sup>3</sup>](https://www.topendsports.com/sport/basketball/anthropometry-nba.htm) who have successfully employed a lineup prioritizing agility and precision shooting over traditional reliance on physically dominant centers. Studies have shown a general decrease in player BMI, indicating a preference for leaner physiques in the league

&emsp; We will dive deeper into these trends, exploring if there is an increasing preference for taller, leaner players in the NBA. We specifically seek to explore the correlation between a player’s height and weight ratio and their on-court performance, considering how these physical attributes contribute to the fast-evolving dynamics of the NBA.

# Hypothesis


- Is there any correlation between players’ height/weight ratio and their draft pick?
    - We predict that taller players with lower weights will be placed higher in the draft pick.

- Is there any correlation between players’ height/weight ratio and their position?
    - Centers and Forwards tend to be the tallest players on the team, as they have responsibilities centered around the basket. They also have to be strong, tending to have high weights to support their confrontations with the other team. 
    - Guards will be shorter in comparison, as they are more known for their agility, speed, and ball dexterity. Their height and weight will be lower. 
- Has there been a statistically significant shift in the average height and weight of NBA players over the last five years?
    - The hypothesis posits that in the last 5 years, there will be no notable shifts in the average height and weight of players for the Guard position, whereas the Forward and Center positions will show an increase in height and decrease in weight. 

**Reasoning:**
- As the pace of NBA games increases, traditional categories of positions are gradually being phased out. For example, positions such as Forward and Center, which have traditionally favored tall and bulky players, are now tending towards a more versatile playstyle of knowing how to shoot as well as dribble up the court. Since there is less of a need for post-work, we expect the traditional image of big stocky centers to be fading away in favor of more mobile tall men.

## Data Wrangling and Cleaning

Here, we used some basic cleaning techniques to prepare the data for analysis. We merged datasets, dropped irrelevant columns, renamed some columns for readability, and modified values for easier manipulation among other techniques. More detailed explanations are provided in the comments below.

```python
# Define function that converts heights to inches
def height_to_inches(height):
    height = str(height)
    height.strip(" ")
    feet, inches = height.split("'", 1)
    inches = inches.replace("''", "")
    return int(feet) *12 + float(inches)
```

```python
# parsed Years 2023 - 2024 -> 2024
df['Year'] = df['Year'].str[:4]
# droped some columns we thought it was irrelevant
cleaned_data = df.drop(columns = ['BODY FAT %', 'HEIGHT W/ SHOES', 'HAND LENGTH (INCHES)', 'HAND WIDTH (INCHES)', 'STANDING REACH'])
# dropped rows that have null data
cleaned_data = cleaned_data.dropna(axis = 0)
# converted WINGSPAN column to inches
cleaned_data['WINGSPAN'] = cleaned_data['WINGSPAN'].apply(height_to_inches)
# converted HEIGHT W/O SHOES column to inches
cleaned_data['HEIGHT W/O SHOES'] = cleaned_data['HEIGHT W/O SHOES'].apply(height_to_inches)
# cleaned Weight column
cleaned_data = cleaned_data.drop(cleaned_data[cleaned_data['WEIGHT (LBS)'] == '-'].index)
cleaned_data['WEIGHT (LBS)'] = cleaned_data['WEIGHT (LBS)'].apply(lambda x: float(x))
# calculate BMI according to weight and height w/o shoes column
cleaned_data['BMI'] = (cleaned_data['WEIGHT (LBS)'] / (cleaned_data['HEIGHT W/O SHOES'])**2) *703
cleaned_data
```

```python
# group position to 3 categories by similar physical traits C, F, G
# SG -> G, PG -> G, SF -> F, PF -> F, C -> C, SG-SF -> G, SF-SG -> G, PG-SG -> G, PF-C -> F, SF-PF -> F, SG-PG -> G, C-PF -> C, PF-SF -> F
# of each player to similar traits
cleaned_data['POS'].unique()
position_mapping = {
    'SG': 'G', 'PG': 'G', 'SF': 'F', 'PF': 'F', 'C': 'C',
    'SG-SF': 'G', 'SF-SG': 'G', 'PG-SG': 'G', 'PF-C': 'F',
    'SF-PF': 'F', 'SG-PG': 'G', 'C-PF': 'C', 'PF-SF': 'F'
}
cleaned_data['POS'] = cleaned_data['POS'].replace(position_mapping)
```

```python
# check if pos got correctly converted
cleaned_data['POS'].unique()
```

```python
# Merging 2 tables
merged = numPickdf.merge(cleaned_data, on = "PLAYER", how='inner')
merged = merged.drop(columns = 'AFFILIATION')
merged = merged.rename(columns = {'ROUND\nNUMBER': 'ROUND', 'ROUND\nPICK': 'PICK', 'OVERALL\nPICK': 'OVERALL_PICK', 'HEIGHT W/O SHOES': 'HEIGHT', 'WEIGHT (LBS)': 'WEIGHT'})
merged
```

```python
# type cast Year column from string to int

merged['Year'] = merged['Year'].astype(int)
```

## 20 years period (height, weight, BMI vs year)

### Mean Height vs Time

```python
# Calculate the mean height for each year
mean_height_by_year = merged.groupby('Year')['HEIGHT'].mean().reset_index()

# Fit the linear regression model
formula = 'HEIGHT ~ Year'
y, X = patsy.dmatrices(formula, data=mean_height_by_year, return_type='dataframe')
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Generate predictions for the regression line
years_range = np.linspace(mean_height_by_year['Year'].min(), mean_height_by_year['Year'].max(), 100)
years_range_df = pd.DataFrame({'Year': years_range})
years_range_df = patsy.dmatrix('Year', data=years_range_df, return_type='dataframe')

predicted_height = model.predict(years_range_df)

# Plot the mean height by year
plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_height_by_year, x='Year', y='HEIGHT', label = 'mean height line')

# Overlay the regression line
plt.plot(years_range, predicted_height, color='red', label='Overall Regression Line')

# Labeling the axes and title
plt.title('Mean Height Over the Years with Regression Line')
plt.ylabel('Height in Inches')
plt.xlabel('Year')
plt.legend()

# Show the plot
plt.show()
```

We observe a decline in player height over the years (negative slope in the Linear Regression Graph). The OLS model indicates a R squared value of approximately 5.6%. P-value of 0.266. Not statistically significant


### Mean Weight vs Time

```python
# Calculate the mean weight for each year
mean_weight_by_year = merged.groupby('Year')['WEIGHT'].mean().reset_index()

# Fit the linear regression model
formula = 'WEIGHT ~ Year'
y, X = patsy.dmatrices(formula, data=mean_weight_by_year, return_type='dataframe')
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Generate predictions for the regression line
predicted_weight = model.predict(years_range_df)

# Plot the mean weight by year
plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_weight_by_year, x='Year', y='WEIGHT', label='Mean Weight Line')

# Overlay the regression line
plt.plot(years_range, predicted_weight, color='red', label='Overall Regression Line')

# Labeling the axes and title
plt.title('Mean Weight Over the Years with Regression Line')
plt.ylabel('Weight in Pounds')
plt.xlabel('Year')
plt.legend()

# Show the plot
plt.show()
```

We observe a decline in player weight over the years (negative slope in the Linear Regression Graph). OLS model indicates a R squared value of approximately 24.6%. P-value of .014. Overall model is statistically significant ( p-value), yet there’s variability ( R squared)  in player weight and year.

We observe a decline in player BMI over the years (negative slope in the Linear Regression Graph). OLS model indicates a R squared value of approximately 28.3%. P-value of .007. ## Independent Samples t-test: NBA physical characteristics analysis: recent 5 years vs previous years

```python
# Filter for recent 5 years
recent5_yrs_merged = merged[merged['Year'].isin(range(2019, 2024))]

# Filter for the remaining years
other_years_merged = merged[~merged['Year'].isin(range(2019, 2024))]


# Conduct the Independent Samples T-test

# height
t_stat, p_value = stats.ttest_ind(recent5_yrs_merged['HEIGHT'], other_years_merged['HEIGHT'], equal_var=False)

print("T-statistic for height:", t_stat)
print("P-value for height:", p_value)

# weight
t_stat, p_value = stats.ttest_ind(recent5_yrs_merged['WEIGHT'], other_years_merged['WEIGHT'], equal_var=False)

print("T-statistic for weight:", t_stat)
print("P-value for weight:", p_value)

# BMI
t_stat, p_value = stats.ttest_ind(recent5_yrs_merged['BMI'], other_years_merged['BMI'], equal_var=False)

print("T-statistic for BMI:", t_stat)
print("P-value for BMI:", p_value)
```

```python
def perform_regression(data, position_name, feature_name):
    X = data[[feature_name]]
    X = sm.add_constant(X)  # Adds a constant term to the predictor
    y = data['OVERALL_PICK']

    model = sm.OLS(y, X).fit()
    print(f"Regression Summary for {position_name} - {feature_name}:")
    print(model.summary())

    return model

def plot_regression(data, position_name, feature_name, model):
    sns.scatterplot(x=feature_name, y='OVERALL_PICK', data=data)
    plt.plot(data[feature_name], model.predict(sm.add_constant(data[feature_name])), color="red")
    plt.title(f"{position_name} - {feature_name} vs Overall Draft Pick")
    plt.xlabel(feature_name)
    plt.ylabel('Overall Draft Pick')
    plt.show()

model_center_height = perform_regression(centers, "Center", "HEIGHT")
plot_regression(centers, "Center", "HEIGHT", model_center_height)
```
